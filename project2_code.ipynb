```python
from ggplot import *
import pandas as pd
from numpy import matrix
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import cross_validation
%matplotlib inline
```

```python
df1 = pd.read_csv('../project_2/data/bank/bank.csv', delimiter=";")
#print df1.count()
df1.head()
```

```python
str_col = [key for key in dict(df1.dtypes) if dict(df1.dtypes)[key] not in ['float64', 'int64']]
```

```python
#look at values of all categorical variables
print str_col
for i in str_col:
    print df1[i].value_counts()
```

```python
#convert yes no to indicator variables
map1 = {'yes': 1.0, 'no': 0.0, '?':np.nan}
df2=df1.replace({'y': map1  \
             ,'loan': map1     \
             ,'housing': map1  \
            })
df2.head()
```

```python
#convert string variables to indicator variables
#step1: convert unknown to column specific unknown
#'education', 'poutcome', 'default', 'marital',  'job', 'contact', 'month'
map_education = {'unknown': 'edu_na', 'primary':'edu_primary', 'secondary':'edu_secondary', 'tertiary':'edu_tertiary'}
map_poutcome  = {'unknown': 'poutcome_na','success':'poutcome_success','failure':'poutcome_failure','other':'poutcome_other'}
map_default   = {'yes': 'default_yes','no':'default_no'}
map_marital   = {'married': 'marital_married','single':'marital_single','divorced':'marital_divorced' }
map_job       = {'unknown': 'job_na'}
map_contact   = {'unknown': 'contact_na', 'cellular':'contact_cellular','telephone':'contact_telephone' }

df3=df2.replace({'education': map_education    \
                ,'poutcome' : map_poutcome     \
                ,'default'  : map_default      \
                ,'marital'  : map_marital      \
                ,'job'      : map_job          \
                ,'contact'  : map_contact      \
                })

df3.head()
```

```python
df3_education = pd.get_dummies(df3['education']) 
df3_poutcome  = pd.get_dummies(df3['poutcome']) 
df3_default   = pd.get_dummies(df3['default']) 
df3_marital   = pd.get_dummies(df3['marital']) 
df3_contact   = pd.get_dummies(df3['contact']) 
df3_job       = pd.get_dummies(df3['job']) 
df3_month     = pd.get_dummies(df3['month']) 

df4 = pd.concat([df3, df3_education,df3_poutcome,df3_default,df3_marital,df3_contact,df3_job,df3_month], axis=1)
df4.head()
```
